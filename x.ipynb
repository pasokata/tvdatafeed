```python
from tvDatafeed import TvDatafeed
import requests
import json

tv = TvDatafeed()
```

```python
text = "BTCUSD"
exchange = "BINANCE"
url = TvDatafeed._TvDatafeed__search_url.format(text, exchange)

# Need Referer _and_ Origin headers, otherwise their Nginx frontend returns 403 :-(
resp = requests.get(url, headers=TvDatafeed._TvDatafeed__search_headers, timeout=60)
resp.raise_for_status()

# As of v3, this returns a dict: {'symbols_remaining':<int>, 'symbols': [...]}
symbols_list = json.loads(resp.text.replace('</em>', '').replace('<em>', '')).get('symbols', [])
```
